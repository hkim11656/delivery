### 4. 2020년 업종별 창업 비율

```python
# 2020년 행만 추출
category_2020 = category1.loc['2020'].to_frame()
category_2020
```

```python
# 2020년 업종별 창업 수 합계
category_sum1 = category_2020['2020'].sum()
category_sum1
```

```python
ratio_category = []

for i in category_2020['2020']:
    ratio_category.append(i/category_sum1*100)

print(ratio_category)
```

```python
# 차트 크기 설정
plt.rcParams["figure.figsize"] = (14,10)

labels_category = category_2020.index.tolist() # 인덱스 리스트로 추출

explode = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0', '#ADD8E6', '#D4DBE1', '#F0F1D0'] 
#colors = ['#FF7AB6', '#7B76FF', '#FFFFC3', '#FFE1A0', '#FFBE91', '#FF999B', '#DF6FDE']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}


plt.pie(ratio_category, labels=labels_category, autopct='%.1f%%', startangle= 90, counterclock=False, explode=explode, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

**=>2020년 업종별 창업 비율을 살펴본 결과, 서비스업이 77.5%로 대부분을 차지하며
